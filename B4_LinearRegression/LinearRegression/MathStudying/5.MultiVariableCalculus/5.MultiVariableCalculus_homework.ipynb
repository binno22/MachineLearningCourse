1. Cho hàm số $L(x, y) = \log(e^x + e^y)$, hãy tính gradient của hàm số $L(x, y)$. Tổng của các thành phần trong véc tơ gradient bằng bao nhiêu?

$$\frac{∂L}{∂x} = \frac{e^x}{e^x + e^y}$$
$$\frac{∂L}{∂y} = \frac{e^y}{e^x + e^y}$$
Vậy 
$$⇒ ∇L(x, y) = \begin{bmatrix} \frac{e^x}{e^x + e^y} \\  \frac{e^y}{e^x + e^y} \end{bmatrix}$$

Tổng các thành phần trong vector gradient bằng $1$.

2. Cho hàm số $f(x, y) = x^2y + y^2x$. Tìm các điểm critical points của hàm số trên.

Ta có:
$$∇f(x, y) = \begin{bmatrix} \frac{∂f}{∂x} \\  \frac{∂f}{∂x} \end{bmatrix} = \begin{bmatrix} 2xy + y^2 \\  x^2 + 2xy \end{bmatrix}$$

Giải $∇f(x,y) = \vec{0}$ ta được:
\begin{align*}
    2xy + y^2 = 0 \\
    x^2 + 2xy = 0 \\
\end{align*}

Trừ 2 vế của phương trình trên ta được
$$y^2 - x^2 = 0 \\
⇒ x = \pm y$$

Thay $x$ vào phương trình $2xy + y^2 = 0$ ta được 
$$x = y = 0$$

3. Mỗi đồ thị tính toán của một mạng neural network như sau:

![](https://imgur.com/dH9q50h.png)

Trong đó:

$$\begin{split}\begin{aligned}f(u, v) & = (u+v)^{2} \\u(a, b) & = (a+b)^{2}, \qquad v(a, b) = (a-b)^{2}, \\a(w, x, y, z) & = (w+x+y+z)^{2}, \qquad b(w, x, y, z) = (w+x-y-z)^2.\end{aligned}\end{split}
$$

Dựa vào công thức Multivariate Chain Rule hãy tính đạo hàm riêng phần $\frac{\delta f}{\delta b}$.

Ta có:
$$\frac{∂f}{∂u} = 2(u+v), \frac{∂f}{∂v} = 2(u+v), \frac{∂u}{∂b} = 2(a+b), \frac{∂v}{∂b} = -2(a-b)$$

Mà theo công thức ta có:
\begin{align*}
    \frac{∂f}{∂b} &= \frac{∂f}{∂u} \frac{∂u}{∂b} + \frac{∂f}{∂v} \frac{∂v}{∂b} \\
    &= 4(u+v)(a+b) - 4(u+v)(a-b) \\
    &= 8b \left( (a+b)^2 + (a-b)^2 \right) \\
    &= 16b (a^2 + b^2)
\end{align*}

Thay $a, b$ vào công thức trên ta được kết quả cuối cùng.

4. Cho hàm số:

$$f(x, y) = xe^{-x^2-y^2}$$

Hãy tính Hessian matrix của hàm số trên. Tìm biểu diễn xấp sỉ của hàm $f(x, y)$ theo Hessian matrix theo quadratic form  

Ta có:
\begin{split} \nabla f(x, y) = e^{-x^2-y^2}\begin{pmatrix}1-2x^2 \\ -2xy\end{pmatrix} \\ 
\mathbf{H}f(x, y) = e^{-x^2-y^2}\begin{pmatrix} 4x^3 - 6x & 4x^2y - 2y \\ 4x^2y-2y &4xy^2-2x\end{pmatrix}.\end{split}

Theo công thức xấp xỉ ta có:
$$ f(\mathbf{x}) = f(\mathbf{x}_0) + \nabla f (\mathbf{x}_0) \cdot (\mathbf{x}-\mathbf{x}_0) + \frac{1}{2}(\mathbf{x}-\mathbf{x}_0)^\top \mathbf{H} f (\mathbf{x}_0) (\mathbf{x}-\mathbf{x}_0) $$

Xấp xỉ $f(x)$ tại $x_0 = (-1, 0)$ ta được 
$$
f(x_0) = -e^{-1} \\
∇f(x_0) = (-e^{-1}, 0) \\
\mathbf{H}f(x_0) = e^{-1} \begin{pmatrix} 2 & 0 \\ 0 & 2 \end{pmatrix}
$$

Từ đó:
$$ f(x, y) \approx e^{-1}\left(-1 - (x+1) +(x+1)^2+y^2\right).$$

Qui ước viết hoa in đậm thì là ma trận, nếu viết thường in đậm thì là véc tơ cột.

5. Với $\mathbf{A}$ là ma trận, $\mathbf{w}$ là véc tơ. Tính: $\nabla_{\mathbf{w}} ||\mathbf{Aw}-\mathbf{y}||^{2}$.

_Gợi ý_: Áp dụng công thức chain rule của matrix-function.


Ta có:
\begin{align*}
     ||\mathbf{Aw}-\mathbf{y}||^{2} &= (\mathbf{Aw}-\mathbf{y})^{\intercal}(\mathbf{Aw} -\mathbf{y}) \\
     &= (\mathbf{w}^{\intercal} \mathbf{A}^{\intercal}  - \mathbf{y}^{\intercal})(\mathbf{Aw} -\mathbf{y}) \\
     &= \mathbf{w}^{\intercal} \mathbf{A}^{\intercal} \mathbf{Aw} - 2 \mathbf{y}^{\intercal} \mathbf{Aw} + \mathbf{y}^{\intercal} \mathbf{y}
\end{align*}

Mà 

$
\nabla_{\mathbf{w}} (\mathbf{w}^{\intercal} \mathbf{A}^{\intercal} \mathbf{Aw}) = (\mathbf{A}^{\intercal} \mathbf{A} + \mathbf{A} \mathbf{A}^{\intercal}) \mathbf{w} \qquad \text{(theo câu 8)} 
$

$\nabla_{\mathbf{w}} (\mathbf{y}^{\intercal} \mathbf{Aw}) = \mathbf{A}^{\intercal} \mathbf{y} $

$\nabla_{\mathbf{w}} (\mathbf{y}^{\intercal} \mathbf{y}) = \mathbf{0}$

Do đó:
$$\nabla_{\mathbf{w}} ||\mathbf{Aw}-\mathbf{y}||^{2} = (\mathbf{A}^{\intercal} \mathbf{A} + \mathbf{A} \mathbf{A}^{\intercal}) \mathbf{w} - 2\mathbf{A}^{\intercal} \mathbf{y}$$


6. $\nabla_{\mathbf{w}}^{2} ||\mathbf{Aw}-\mathbf{y}||^{2}$.

Theo câu 5, ta có:
$$\nabla^2_{\mathbf{w}} ||\mathbf{Aw}-\mathbf{y}||^{2} = \nabla_{\mathbf{w}} \left( (\mathbf{A}^{\intercal} \mathbf{A} + \mathbf{A} \mathbf{A}^{\intercal}) \mathbf{w} - 2\mathbf{A}^{\intercal} \mathbf{y} \right) = \mathbf{A}^{\intercal} \mathbf{A} + \mathbf{A} \mathbf{A}^{\intercal}$$

7. Tính: $\nabla_{\mathbf{x}}\mathbf{a^{\intercal}\mathbf{x}\mathbf{x}^{\intercal}\mathbf{b}}$

_Gợi ý_: Sử dụng product rule.


**Product rule**
$$\nabla\left( f(\mathbf{X})^Tg(\mathbf{X}) \right) = \left(\nabla f(\mathbf{X})\right) g(\mathbf{X}) + \left(\nabla g(\mathbf{X})\right) f(\mathbf{X})$$

Ta có: 

\begin{align*}
    \nabla_{\mathbf{x}} \mathbf{a^{\intercal}\mathbf{x}\mathbf{x}^{\intercal}\mathbf{b}} &= \nabla_{\mathbf{x}} (\mathbf{a^{\intercal}\mathbf{x}) (\mathbf{x}^{\intercal}\mathbf{b}}) \\
    &= \mathbf{a} \mathbf{x}^{\intercal}\mathbf{b} + \mathbf{ba}^{\intercal} \mathbf{x} \\
    &= \mathbf{ab}^{\intercal} \mathbf{x} + \mathbf{ba}^{\intercal} \mathbf{x}
\end{align*}

8. Tính: $\nabla_{\mathbf{x}}\mathbf{x}^{\intercal}\mathbf{A}\mathbf{x}$ 

_Gợi ý_: Sử dụng product rule.


\begin{align*} 
    \nabla_{\mathbf{x}} \mathbf{x}^{\intercal}\mathbf{A}\mathbf{x} &= \nabla_{\mathbf{x}} (\mathbf{x}^{\intercal}) (\mathbf{A}\mathbf{x}) \\
    &= \mathbf{IA} \mathbf{x} + \mathbf{A}^{\intercal} \mathbf{x} \\
    &= (\mathbf{A} + \mathbf{A}^{\intercal}) \mathbf{x}
\end{align*} 

Có thể tham khảo cách chứng minh một cách chi tiết [tại đây](https://www.cs.ubc.ca/~schmidtm/Courses/340-F16/linearQuadraticGradients.pdf).

9. Tính: $\nabla_{\mathbf{x}}(\mathbf{A}\mathbf{x} + \mathbf{b})^{\intercal}\mathbf{C}(\mathbf{D}\mathbf{x} + \mathbf{e})$ 

_Gợi ý_: Sử dụng product rule.


\begin{align*}
    \nabla_{\mathbf{x}}(\mathbf{A}\mathbf{x} + \mathbf{b})^{\intercal}\mathbf{C}(\mathbf{D}\mathbf{x} + \mathbf{e}) &= \nabla_{\mathbf{x}} (\mathbf{A}\mathbf{x} + \mathbf{b})^{\intercal} (\mathbf{CD}\mathbf{x} + \mathbf{Ce}) \\
    &= \mathbf{A}^{\intercal}(\mathbf{CD}\mathbf{x} + \mathbf{Ce}) + ( \mathbf{CD})^{\intercal} (\mathbf{A}\mathbf{x} + \mathbf{b})
\end{align*}

10. Chứng minh: $$\nabla_{\mathbf{A}} ~ \text{trace}(\mathbf{AB}) = \nabla_{\mathbf{A}} ~ \text{trace}(\mathbf{BA})  = \mathbf{B}^{\intercal}$$

_Gợi ý_: Chứng minh phần tử $a_{ij}$ ở hai vế bằng nhau.

Trong bài tập các tuần trước, ta đã chứng minh được $\text{trace}(\mathbf{AB}) = \text{trace}(\mathbf{BA}) $. Nên giờ ta chỉ cần chứng minh 
$$\nabla_{\mathbf{A}} \text{trace}(\mathbf{AB}) = \mathbf{B}^{\intercal}$$

Chứng minh:
\begin{align*}
    \text{trace}(\mathbf{AB}) &= (\mathbf{AB}) _{11} + (\mathbf{AB}) _{22} + \cdots (\mathbf{AB}) _{nn} \\
    &= \sum_{j=1}^n a_{1j} b_{j1} + \sum_{j=1}^n a_{2j} b_{j2} + \cdots + \sum_{j=1}^n a_{nj} b_{jn} 
\end{align*}

$
⇒ \frac{\partial \text{trace}(\mathbf{AB})}{\partial a_{ij}} = b_{ji} \\
⇒ \nabla_{\mathbf{A}} \text{trace}(\mathbf{AB}) = \mathbf{B}^{\intercal}
$
