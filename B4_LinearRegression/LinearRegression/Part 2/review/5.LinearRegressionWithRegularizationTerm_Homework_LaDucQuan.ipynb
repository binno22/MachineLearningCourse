```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test_2.csv')
print(df_train.shape)
print(df_test.shape)
```

Đối với file train, có 211 cột, file test có 147 cột: 
- 1 cột Id
- 25 cột Feature_1 tới Feature_25: các features khác có liên quan tới việc dự báo
- 1 cột Ret_MinusTwo: chênh lệch về giá đóng phiên của ngày D-2 so với ngày D-1 (1 ngày)
- 1 cột Ret_MinusOne: chênh lệch về giá đóng phiên của ngày D-1 so với thời điểm giao dịch trong ngày bắt đầu (1/2 ngày)
- 179 cột Ret_2 tới Ret_180: chênh lệch về giá giữa một phút và phút trước đó trong ngày giao dịch. (đối với file test thì bỏ 60 cột từ Ret 121 tới 180).
- 1 cột Ret_PlusOne: chênh lệch về giá giữa thời điểm Ret_180 của ngày D với thời điểm đóng phiên ngày D+1 (1 ngày, đối với file test bỏ cột này)
- 1 cột Ret_PlusTwo: chênh lệch về giá giữa thời điểm đóng phiên ngày D+1 với thời điểm đóng phiên ngày D+2 (1 ngày, đối với file test bỏ cột này)
- 1 cột Weight_Intraday: trọng số dùng để đánh giá dự đoán trong ngày, các dự đoán từ phút 121 tới 180 (file test bỏ cột này)
- 1 cột Weight_Daily: trọng số dùng để đánh giá dự đoán ngày tiếp theo, các dự đoán ngày D+1 và D+2 (file test bỏ cột này)

```python
features_variables = ['Feature_{}'.format(i) for i in range(1,26)] #các biến có thể dùng hoặc không dùng
return_variables_X = ['Ret_MinusTwo','Ret_MinusOne']+['Ret_{}'.format(i) for i in range(2,121)] #các biến features
return_variables_Y = ['Ret_{}'.format(i) for i in range(121,181)] + ['Ret_PlusOne','Ret_PlusTwo'] #các biến target
```

```python
#thay các giá trị missing trong các DÒNG bằng trung bình của DÒNG đó, vì mỗi dòng là một cổ phiếu khác nhau

TRAINING_DATA = df_train[return_variables_X+return_variables_Y]
TRAINING_DATA = TRAINING_DATA.T.fillna(TRAINING_DATA.mean(axis=1)).T

TESTING_DATA = df_test[return_variables_X]
TESTING_DATA = TESTING_DATA.T.fillna(TESTING_DATA.mean(axis=1)).T
```

```python
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from tqdm.notebook import tqdm
```

## Dự đoán không sử dụng 25 cột feature

Do các giá trị dự đoán là rất nhỏ (< 1) nên sử dụng đánh giá là MAPE

```python
lr = []
for i in range(62):
  lr = lr + [LinearRegression()]
```

### Dự đoán phút thứ 121

lr[0] sẽ tương ứng với Ret_121 <br>
lr[1] tương ứng Ret_122 <br>
... <br>
lr[59] tương ứng Ret_180 <br>
lr[60] tương ứng Ret_PlusOne <br>
lr[61] tương ứng Ret_PlusTwo <br>

```python
X_train = TRAINING_DATA[return_variables_X]
Y_train = TRAINING_DATA['Ret_121']

x_train, x_val, y_train, y_val = train_test_split(X_train, Y_train, random_state = 42, test_size = 0.2)

lr[0].fit(x_train, y_train)
print('MAPE of train set:',mean_absolute_percentage_error(lr[0].predict(x_train), y_train))
print('MAPE of validation set:',mean_absolute_percentage_error(lr[0].predict(x_val), y_val))
```

### Dự đoán tất cả các kết quả nhưng chỉ dựa vào dữ liệu có sẵn

Xây dựng 61 model dựa đoán các giá trị từ ret_121 đến ret_180 và ret_plusone, ret_plustwo nhưng chỉ dựa vào 121 biến ban đầu

```python
X_train = TRAINING_DATA[return_variables_X] #X_train không thay đổi 
mape_train_1 = []
mape_val_1 = []

for i in range(len(return_variables_Y)):
  Y_train = TRAINING_DATA[return_variables_Y[i]]
  x_train, x_val, y_train, y_val = train_test_split(X_train, Y_train, random_state = 42, test_size = 0.2)
  lr[i].fit(x_train, y_train)
  mape_train_1.append(mean_absolute_percentage_error(lr[i].predict(x_train), y_train))
  mape_val_1.append(mean_absolute_percentage_error(lr[i].predict(x_val), y_val))

print('Average MAPE for training set:',np.mean(mape_train_1))
print('Average MAPE for validation set:',np.mean(mape_val_1))
```

### Dự đoán các thời điểm sau này với dữ liệu từ dữ đoán của thời điểm trước

Xây dựng 61 model dự đoán các giá trị, với biến đầu vào của mô hình sau sẽ gồm dữ liệu đầu vào đã cho và kết quả dự đoán của các mô hình đi trước

```python
X_train = TRAINING_DATA[return_variables_X] #X_train ban đầu 
mape_train_2 = []
mape_val_2 = []

for i in range(len(return_variables_Y)):
  Y_train = TRAINING_DATA[return_variables_Y[i]]
  x_train, x_val, y_train, y_val = train_test_split(X_train, Y_train, random_state = 42, test_size = 0.2)
  lr[i].fit(x_train, y_train)
  mape_train_2.append(mean_absolute_percentage_error(lr[i].predict(x_train), y_train))
  mape_val_2.append(mean_absolute_percentage_error(lr[i].predict(x_val), y_val))
  X_train.insert(121+i, return_variables_Y[i], lr[i].predict(X_train))  #thêm kết quả dự đoán vào X_train, chuẩn bị cho mô hình tiếp theo

print('Average MAPE for training set:',np.mean(mape_train_2))
print('Average MAPE for validation set:',np.mean(mape_val_2))
```

Có thể thấy cách thức dự đáon với dữ liệu từ thời điểm trước cho ra kết quả tốt hơn (MAPE nhỏ hơn). Với mô hình xây dựng theo cách 2, kết quả bị overfit rất nặng. 

Khi thử trên dữ liệu đầy đủ (tất cả các cột đều không có missing value), kết quả dự đoán tốt hơn, chỉ dừng ở mức 10%

### Dự đoán test set với mô hình xây dựng theo cách đầu tiên

```python
#Xây dựng mô hình
X_train = TRAINING_DATA[return_variables_X] 
X_test = TESTING_DATA[return_variables_X]
Y_hat = {}
#Y_hat là ma trận kết quả
for i in range(len(return_variables_Y)):
  Y_train = TRAINING_DATA[return_variables_Y[i]]
  lr[i].fit(X_train, Y_train)
  Y_hat[return_variables_Y[i]] = lr[i].predict(X_test)

Y_hat = pd.DataFrame(Y_hat)
print('5 dòng đầu của ma trận kết quả dự đoán')
Y_hat.head()
```
