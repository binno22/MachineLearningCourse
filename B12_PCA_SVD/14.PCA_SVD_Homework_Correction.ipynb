```python
import numpy as np
import pandas as pd
import seaborn as sns

from matplotlib import pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
```

```python
# 1
# Load dataset
X, y = load_breast_cancer(return_X_y=True, as_frame=True)
X.head()
```

1. Giảm chiều dữ liệu bằng thuật toán SVD

```python
def SVDDecomposition(X):
    """
    Argument:
        X: a matrix of dataset with dimension (n, m)
        
    Returns:
        U: a matrix of left singular vectors with dimesion (n, n)
        S: a matrix singular values with dimesion (n, m)
        Vt: a transposed matrix of left singular vectors with dimesion (m, m)
    """
    n, m = X.shape[0], X.shape[1]
    U, S, Vt = np.linalg.svd(X)
    
    # Check some properties
    assert U.shape == (n, n), 'Invalid shape'
    # Because every value of S equal 0, except for major diagonal 
    # To reduce storage, numpy converts S from (n, m) to (m,)
    assert S.shape == (m,), 'Invalid shape'
    assert Vt.shape == (m, m), 'Invalid shape'
    assert np.all(S > 0), 'Invalid Sigma'
    
    return U, S, Vt

def project(X, d):
    """
    Arg: 
        X: a matrix of dataset with dimension (n, m)
        d: number of components that you need
        
    Return:
        Xd: a matrix, projection of X on d-dimension space
    """
    # get svd decomposition 
    U, S, Vt = SVDDecomposition(X)
    
    # project down to d dimensions
    Wd = Vt.T[:, :d]
    Xd = X.dot(Wd)
    
    return - Xd

# Test function
scaler = StandardScaler()
X_norm = scaler.fit_transform(X)

d = 2
Xd = project(X_norm, d)
Xd
```

2. Giảm chiều bằng thuật toán PCA

**Cách 1:** Code theo các bước của thuật toán

```python
from numpy import mean
from numpy import cov, array
from numpy.linalg import eig

def compute_PCA(X, d):
    """
    Argument:
        X: a matrix of dataset with dimension (n, m)
        d: number of compressed dimensions that we need to achive.
    Returns:
        U: a matrix with dimesion (n, d)
    """
    # Step 1: Standard normalization X (with each feature x_std = (x-mean(x))/std(x)) with shape (m, n)
    X_std = (X - X.mean(axis=0))/X.std(axis=0)
    # Step 2: Calculate covariance matrix of centered matrix
    V = cov(X_std.T)
    # Step 3: Compute eigenvalues and eigenvectors of covariance matrix
    values, vectors = eig(V)
    # Step 4: Select d eigenvectors corresponding with top d highest eigenvalues:
    top_k_indices = np.argsort(values, axis=0)[::-1][:d]
    # Step 5: Project data into top d eigenvectors
    X_pca = X_std.dot(vectors[:, top_k_indices]).values
    return X_pca

X_pca2 = compute_PCA(X=X, d=2)
X_pca2
```

Cách 2: Giảm chiều bằng PCA class của sklearn package

```python
# 2
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_norm)
X_pca
```

```python
# 3. Visualization

labels = np.unique(y)

for label in labels:
    ids = np.where(y == label)[0]
    plt.scatter(X_pca[ids, 0], X_pca[ids, 1], label=label)

plt.title('PCA reduces 2D dimensions')
plt.legend()
```

```python
# 3. Visualization

labels = np.unique(y)

for label in labels:
    ids = np.where(y == label)[0]
    plt.scatter(X_pca2[ids, 0], X_pca2[ids, 1], label=label)

plt.title('PCA reduces 2D dimensions coding yourself')
plt.legend()
```
