````python
labels = []
documents = []

for root, dirs, files in os.walk('Test_Full'):
  for file in files:
    if '.txt' in file:
      label = os.path.basename(root)
      with open(os.path.join(root, file), 'r', encoding='utf-16') as f:
        document = f.read()
      labels.append(label)
      documents.append(document)

print('total documents: ', len(documents))
print('summary labels: ', np.unique(labels, return_counts = True))
````

````python
# To reduce computational cost, we take on each topics 100 documents.
sub_labels = []
sub_documents = []
dict_counts = {
    'Chinh tri Xa hoi': 0,
    'Doi song': 0,
    'Khoa hoc': 0,
    'Kinh doanh': 0,
    'Phap luat': 0,
    'Suc khoe': 0,
    'The gioi': 0,
    'The thao': 0,
    'Van hoa': 0,
    'Vi tinh': 0
}

for i, (label, document) in enumerate(zip(labels, documents)):
  dict_counts[label] += 1
  if dict_counts[label] <= 100:
    sub_labels.append(label)
    sub_documents.append(document)
````

````python
print('total documents: ', len(sub_documents))
print('summary labels: ', np.unique(sub_labels, return_counts = True))
````

````python
# Download stopwords
# !wget https://raw.githubusercontent.com/stopwords/vietnamese-stopwords/master/vietnamese-stopwords.txt

import nltk
from nltk.tokenize import RegexpTokenizer
from nltk.stem import WordNetLemmatizer,PorterStemmer
from nltk.corpus import stopwords
import re
import string
lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer() 

with open('vietnamese-stopwords.txt', 'r', encoding='utf-8') as f:
  stop_words = f.read()
stop_words = stop_words.rstrip().split('\n')

def preprocess(sentence):
  # convert to string
  sentence=str(sentence)
  # convert to lowercase
  sentence = sentence.lower()
  # create tokens
  tokens = word_tokenize(sentence)
  # remove punctuation from each word
  table = str.maketrans('', '', string.punctuation)
  stripped = [w.translate(table) for w in tokens]
  # remove tokens that are not alphabetic
  words = [word for word in stripped if word.isalpha()]
  # remove stopwords
  filtered_words = [w for w in words if len(w) > 2 if not w in stop_words]
  return " ".join(filtered_words)

y = preprocess(sub_documents[0])
y
````

````python
sub_documents_clean=list(map(lambda t: preprocess(t), sub_documents))
````

## Model training

Sử dụng CountVectorizer để đếm tần suất tokens trong văn bản

````python
from sklearn.feature_extraction.text import CountVectorizer
count_vector = CountVectorizer(sub_documents_clean)
print(count_vector)
````

````python
count_vector.fit(sub_documents_clean)
# count_vector.get_feature_names()[-100:]
````

````python
doc_array = count_vector.transform(sub_documents_clean).toarray()
doc_array.shape
````

````python
label_maps = {
    'Chinh tri Xa hoi': 0,
    'Doi song': 1,
    'Khoa hoc': 2,
    'Kinh doanh': 3,
    'Phap luat': 4,
    'Suc khoe': 5,
    'The gioi': 6,
    'The thao': 7,
    'Van hoa': 8,
    'Vi tinh': 9
}

y = [label_maps[label] for label in sub_labels]
y[:5]
````

````python
# Model train
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model.fit(doc_array, y)

# Test accuracy
y_pred = model.predict(doc_array)
print('Accuracy :', np.sum(y_pred==y)/len(y))
````

````python
# Summary result
from sklearn.metrics import classification_report
print(classification_report(y, y_pred))
````

````python
from sklearn.datasets import make_classification

# Khởi tạo dữ liệu example
X, y = make_classification(n_samples=500, n_features=50, random_state=123)
````

````python
from keras.layers import Dense, Input
from keras.models import Model, Sequential
from keras.optimizers import RMSprop, Adam

def autoencoder(input_unit, hidden_unit):
    model = Sequential()
    model.add(Dense(input_unit, input_shape = (15,), activation = 'relu'))
    model.add(Dense(hidden_unit, activation = 'relu'))
    model.add(Dense(input_unit, activation = 'softmax'))
    model.compile(loss = 'categorical_crossentropy', optimizer = Adam(),
                 metrics = ['accuracy'])
    model.summary()
    return model

model_auto = autoencoder(input_unit = 15, hidden_unit = 6)

model_auto.fit(X, X, epochs = 5, batch_size = 3)
````
