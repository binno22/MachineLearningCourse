```python
from sklearn.preprocessing import KBinsDiscretizer
est = KBinsDiscretizer(n_bins=6, encode='ordinal', strategy='uniform')
df['age'] = est.fit_transform(df[['age']]) 
```

```python
def add_working_age(df):
    col_add = []
    for i in range(len(df)):
        if df.loc[i, 'age'] in [1,2,3] and df.loc[i, 'status'] in ['Husband', 'Wife']:
            col_add.append(1)
        else:
            col_add.append(0)
    return col_add

df['working_age'] = add_working_age(df)
```

```python
df.drop(columns=['final_weight', 'ID'], inplace= True)
```

```python
print("INCOME = 1")
for i, feature in enumerate(['capital_loss', 'capital_gain']):
    print(feature, ": ", income1[income1[feature] != 0][feature].mean(), 
              "+/-", income1[income1[feature] != 0][feature].std())
print('\nINCOME = 0')  
for i, feature in enumerate(['capital_loss', 'capital_gain']):
    print(feature, ": ", income0[income0[feature] != 0][feature].mean(), 
              "+/-", income0[income0[feature] != 0][feature].std())
```

```python
def add_profit(df):
    col_add = []
    for i in range(len(df)):
        if df.loc[i, 'capital_loss']>=1800 or df.loc[i, 'capital_gain']>=7000:
            col_add.append(1)
        else:
            col_add.append(0)
    return col_add

df['profit'] = add_profit(df)
```

```python
est = KBinsDiscretizer(n_bins=5, encode='ordinal', strategy='uniform')
df['hrs_per_week'] = est.fit_transform(df[['hrs_per_week']])
```

```python
def add_ability_work(df):
    col_add = []
    for i in range(1, len(df)):
        if df.loc[i, 'total_education_yrs']>=9 and df.loc[i, 'hrs_per_week']>=1:
            col_add.append(1)
        else:
            col_add.append(0)
    return col_add

df['ability_work'] = add_working_age(df)
```

```python
def add_rate_income(df):
    col_add = []
    for i in range(len(df)):
        col_add.append((df.loc[i, 'working_age']+df.loc[i, 'profit']+df.loc[i, 'ability_work']+df.loc[i, 'condition'])/4)
    return col_add

df['rate_high_income'] = add_rate_income(df)
```
