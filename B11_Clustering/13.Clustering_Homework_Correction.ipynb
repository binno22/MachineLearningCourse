Câu 2: Huấn luyện thuật toán K-means để phân cụm các loại rượu phù hợp. Sử dụng phương pháp Elbow để đưa ra số cụm phù hợp cho mô hình

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Normalize
scale = StandardScaler()
X_norm = scale.fit_transform(X)

# Define hyperparameter
n_clusters = 3
seed = 42

kmeans = KMeans(n_clusters=n_clusters, random_state=seed)
kmeans.fit(X_norm)
```

```python
kmeans.labels_, y
```

```python
# Elbow 
from scipy.spatial.distance import cdist

losses = []
K = 10
for i in range(1, K):
    # 1.  Huấn luyện với số cụm = i
    kmeans_i = KMeans(n_clusters=i, random_state=0)
    kmeans_i.fit(X_norm)
    # 2. Tính _hàm biến dạng_
    # 2.1. Khoảng cách tới toàn bộ centroids
    d2centroids = cdist(X_norm, kmeans_i.cluster_centers_, 'euclidean') # shape (n, k)
    # 2.2. Khoảng cách tới centroid gần nhất
    min_distance = np.min(d2centroids, axis=1) # shape (n)
    loss = np.sum(min_distance)
    losses.append(loss)
```

```python
plt.figure(figsize=(12, 8))
plt.plot(np.arange(1, K), losses, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('Distortion')
plt.title('The Elbow Method using Distortion')
plt.show()
```

Câu 3: Giảm chiều dữ liệu ở câu 1. Sử dụng một thuật toán thuộc Phân cụm phân cấp để phân loại các loại rượu. Sau đó visualize kết quả.

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.manifold import TSNE
import scipy.cluster.hierarchy as shc
import matplotlib.patheffects as PathEffects
```

```python
plt.figure(figsize=(20, 7))
plt.title("Wine Dendograms")
dend = shc.dendrogram(shc.linkage(X_norm, method='ward'))
plt.axhline(200, linestyle='--')
plt.xlabel('sample indice')
plt.ylabel('dissimilarity metric cluster')
```

```python
# Biến đổi dữ liệu về 2 chiều
X_tsne = TSNE(n_components=2, random_state=123).fit_transform(X_norm)
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
labels = cluster.fit_predict(X_tsne)
```

```python
def _plot_kmean_scatter(X, labels):
    '''
    X: dữ liệu đầu vào
    labels: nhãn dự báo
    '''
    # lựa chọn màu sắc
    num_classes = len(np.unique(labels))
    palette = np.array(sns.color_palette("hls", num_classes))

    # vẽ biểu đồ scatter
    fig = plt.figure(figsize=(12, 8))
    ax = plt.subplot()
    sc = ax.scatter(X[:,0], X[:,1], lw=0, s=40, c=palette[labels.astype(np.int)])

    # thêm nhãn cho mỗi cluster
    txts = []

    for i in range(num_classes):
        # Vẽ text tên cụm tại trung vị của mỗi cụm
        xtext, ytext = np.median(X[labels == i, :], axis=0)
        txt = ax.text(xtext, ytext, str(i), fontsize=24)
        txt.set_path_effects([
            PathEffects.Stroke(linewidth=5, foreground="w"),
            PathEffects.Normal()])
        txts.append(txt)
    plt.title('t-sne visualization')

_plot_kmean_scatter(X_tsne, labels)
```
